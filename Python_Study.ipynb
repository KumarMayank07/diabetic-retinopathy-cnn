```python
#find the sum of even and odd numbers:

lst=[1,2,3,4,5,6,7]
even_sum=0
odd_sum=0
for i in lst:
  if(i%2==0):
    even_sum+=i
  else:
    odd_sum+=i
print(even_sum)
print(odd_sum)
print("Even Sum is {}".format(even_sum))
print("Odd Sum is {}".format(odd_sum))
```

```python
#While Loop

even_sum=0
odd_sum=0
i=0

while(i<=10):
  if(i%2==0):
    even_sum+=i
  else:
    odd_sum+=i
  i+=1
print(even_sum)
print(odd_sum)
print("Even Sum is {}".format(even_sum))
print("Odd Sum is {}".format(odd_sum))
```

```python
#List:
'''
A list is a data structure in Python that is a mutable, or changeable,
ordered sequence of elements.
Each element or value that is inside of a list is called an item.
Just as strings are defined as characters between quotes,
lists are defined by having values between square brackets [ ]
'''
```

```python
#strings are immutable : if i define the value of string then i cant change its item
str="KUMAR"
print(str)
str="MAYANK"
print(str)
str[0]="K"
print(str)
```

```python
##Lists is mutable
lst=[1,2,3,4,"K","Hello"]
print(lst)
print(lst[4])
lst[4]="M"
print(lst[4])
lst=list((1,2,3,4,5))
print(type(lst))
print(lst)
lst=[1,2,3,4,5]
for i in lst:
    print(i**2)
print(len(lst))
print(max(lst))
print(min(lst))

#.append()
lst=[1, 2, 3, 4, 5]
#.append is used to add elements at the end of the list
lst.append("Kumar")
lst.append(["Kumar","Mayank"])
print(lst[:])
print(lst[6][1])
print(lst[2:6])
print(lst)

#.insert()-->insert in a specific order
lst.insert(2,"Hero")
print(lst)
lst.append(["Hello","World"])
print(lst)
lst=[1,2,3]
lst.append([4,5])
print(lst)

#.extend()--->seperately add element at the end of the list
lst=[1,2,3,4,5,6]
lst.extend([8,9])
print(lst)

#Various Operations that we can perform in List
lst=[1,2,3,4,5]
print(sum(lst))
print(lst*5)#here list have been appended 5 times


#Pop() Method: updates list
lst=[1,2,3,4,5]
lst.pop()
print(lst)
lst.pop(2)
print(lst)

#count():Calculates total occurrence of given element of List
lst=[1,1,2,3,4,5]
print(lst.count(1))
#length:Calculates total length of List
print(len(lst))
# index(): Returns the index of first occurrence. Start and End index are not necessary parameters
print(lst.index(1,1,4))
print(lst.index(1))
##Min and Max
print(min(lst))
print(max(lst))
```

```python
#sets:
'''
data type that is iterable, mutable, and has no duplicate elements.
Python's set class represents the mathematical notion of a set.
This is based on a data structure known as a hash table
'''

print(type({})) #dic has key value pair
print(type({1,2,3,4,5,6,"k"})) #sets has indep elements with no duplicacy
print(type(set([1,2,3,4,5,6,"k"])))#set()-->fn accepts only 1 arg so pass the list
print(set([1,2,3,4,5,6,6,"k"]))


## Defining an empy set

set_var= set()
print(set_var)
print(type(set_var))
set_var={1,2,3,4,3}
print(set_var)
set_var={"Avengers","IronMan",'Hitman'}
print(set_var)
print(type(set_var))

## Inbuilt function in sets
set_var.add("Hulk")
print(set_var)
set1={"Avengers","IronMan",'Hitman'}
set2={"Avengers","IronMan",'Hitman','Hulk2'}
print(set2.intersection(set1))
print(set2)
print(set2.intersection_update(set1))
print(set2)

##Difference
set1={"Avengers","IronMan",'Hitman'}
set2={"Avengers","IronMan",'Hitman','Hulk2'}
print(set2.difference(set1))
print(set2)

## Difference update
print(set2.difference_update(set1))
print(set2)
set1={"Avengers","IronMan",'Hitman'}
set2={"Avengers","IronMan",'Hitman','Hulk2'}
#.issuperset
print(set2.issuperset(set1))
print(set1.issuperset(set2))
print(set1.issubset(set2))
```

```python
#Dictonaries
'''
A dictionary is a collection which is unordered, changeable and indexed.
In Python dictionaries are written with curly brackets, and they have keys and values.
'''


dic={}
print(type(dic))
print(type({}))
print(type({1:"HI"}))
print(type(dict()))
print(dict(name="Kumar",age="21"))#keys should be without quotes
set1={1,2,3,4,5}
print(type(set1))

## Let create a dictionary
my_dict={"Car1": "Audi", "Car2":"BMW","Car3":"Mercidies Benz"}
print(type(my_dict))
print(my_dict)

##Access the item values based on keys
print(my_dict["Car1"])

print(my_dict.keys())
print(my_dict.values())
print(my_dict.items())


# We can even loop throught the dictionaries keys
for x in my_dict: #or my_dict.keys()
    print(x)


# We can even loop throught the dictionaries values
for x in my_dict.values():
    print(x)

# We can also check both keys and values
for x in my_dict.items():
    print(x)

print(my_dict)


## Adding items in Dictionaries
my_dict['car4']='Audi 2.0'
print(my_dict)
my_dict['Car1']='Maruti'
print(my_dict)


#Nested Dictionary
car1_model={'Mercedes':1960}
car2_model={'Audi':1970}
car3_model={'Ambassador':1980}

car_type={'car1':car1_model,'car2':car2_model,'car3':car3_model}
print(car_type)

## Accessing the items in the dictionary
print(car_type['car1'])
print(car_type['car1']['Mercedes'])
```

```python
#Python Numpy Arrays
'''
numpy is a general-purpose array-processing package.
It provides a high-performance multidimensional array object, and tools for working with these arrays.
It is the fundamental package for scientific computing with Python
'''
'''
What is an array?
An array is a data structure that stores values of same data type.
In Python, this is the main difference between arrays and lists.
While python lists can contain values corresponding to different data types, arrays in python can only contain values corresponding to same data type
'''
#import the library
import numpy as np
my_lst=[1,2,3,4,5]
arr=np.array(my_lst)
print(arr)
print(type(arr))
print(arr.shape)

lst1=[1,2,3,4,5]
lst2=[2,3,4,5,6]
lst3=[3,4,5,6,7]
arr1=np.array([lst1,lst2,lst3])
print(arr1)
print(arr1.shape) #(row,coloumn)

print(arr)
#indexing
print(arr[3])
arr[3]=5
print(arr)
print(arr[1:3])
print(arr[-1])
print(arr[:-1])
print(arr[::-2])#last se 2 element ka jump krke print krta h
print(arr[::-3])
print(arr1)
print(arr1[:,1])#select all rows and 1st coloumn
print(arr1[:,1:2])
print(arr1[1:3,1:3])
print(arr1[:,3:])
print(arr1[:,3:].shape)
print(arr1[:,1])
print(arr1[1:,1:3])
print(arr1[1:,3:])


##EDA
print(arr)
print(arr<2)
print(arr[arr<2])
print(arr1)
print(arr1.reshape(5,3))

##mechanism to create an array
print(np.arange(1,20,2).reshape(2,5))#step size=2
print(np.arange(1,20,2).reshape(1,10))
print(np.arange(1,20,2).reshape(2,5,1))
print(arr *arr)
print(arr1 * arr1)
print(np.ones((5,3)))
print(np.zeros((4,5)))
print(np.random.randint(10,50))
print(np.random.randint(10,50,4))
print(np.random.randint(10,50,4).reshape(2,2))
print(np.random.randn(5,6)) #standard normal distribution: mean=0 sd=1
print(np.random.random_sample((4,7)))


```
